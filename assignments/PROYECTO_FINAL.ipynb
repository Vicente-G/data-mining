# Definición del problema

En base a nuestra motivación en donde mencionamos que para clasificar y trabajar de la mejor forma con la información que tenemos necesitamos ayudarnos computacionalmente además de optimizar dichos procesos. Podemos mencionar y darnos cuenta de las dificultades y problemáticas que se nos presentaran para este trabajo.

El problema principal para nuestro trabajo es la `clasificación` y `agrupación` de estrellas en función a sus características espectrales en las siguientes regiones (les llamamos regiones ya que nos basaremos en el diagrama de Hertzsprung-Russell para guiarnos en la explicación) pues en él se observará de manera mucho más clara y fácil el como se distribuyen los diferentes tipos de estrellas según su naturaleza y características. Adjuntamos imagen del diagrama:

image.png

**Supergigantes:** Son estrellas muy luminosas y enormemente grandes que tienen coloración rojiza o amarilla. En su máxima extensión pueden llegar a alcanzar más de mil veces el radio solar. Estas ocupan la parte superior del diagrama de Hertzsprung-Russell. En la clasificación espectral de Yerkes son de clase de luminosidad. Corresponden a las clases de luminosidad Ia o Ib.

**Gigantes:** Son estrellas de coloración rojiza más luminosas. Ocupan la parte superior del diagrama de Hertzsprung-Russell. Corresponden a las clases de luminosidad II y III.

**Secuencia Principal:** Se denomina secuencia principal a la región del diagrama HR en la que las estrellas pasan la mayor parte de su existencia (que técnicamente corresponde a un estadio en el que en la estrella consume hidrógeno). Muchas estrellas caen sobre esta diagonal que va desde el extremo superior izquierdo desde las estrellas muy luminosas blanco-azuladas, hasta el extremo inferior derecho donde están las enanas rojas.

**Enanas blancas:** Son estrellas pequeñas y calientes que agotaron todo el hidrógeno que utilizaban como combustible nuclear, pero de temperaturas superficiales altas, de ahí su coloración blanca. 

Esto lo lograremos con ayuda de la información que tenemos a nuestra disposición la cual en algunos casos está dada y en otros deberemos buscarla o calcularla dependiendo lo que se necesite y en base tambien a las hipotesis que se tengan y las conclusiones esperadas.

Es por esto que para resolver este problema primero pensamos y analizamos sobre la existencia de las distintas formas de clasificación de estrellas, ya sean para las mencionadas anteriormente u otras. Teniendo claro esto y una vez realizado ese análisis nos damos cuenta que nuestra mejor herramienta para este caso será la utilización del `diagrama de Hertzsprung-Russell` (diagrama HR) el cual es un diagrama de dispersión de estrellas en dos dimensiones (es decir para hacer su grafica se necesitan 2 variables) al cual se le pueden dar múltiples usos y que nos ayudara a ver de forma mucho más general y teórica la información. Para este caso en particular lo utilizaremos con la temperatura y luminosidad (ambas con sus respectivas clases) de las estrellas los cuales representaremos en el diagrama HR. Otras razones útiles para la utilización de este diagrama en este caso es que sirve para tratar y trabajar con un gran número de estrellas, presentar de forma sintética diversos resultados y nos ayuda a poder identificar relaciones que nos serán utilices para poder comparar, clasificar y agrupar en los tipos de estrellas ya mencionados de manera precisa.

## Reparación del dataframe

Observamos que el índice del `DataFrame` se encuentra duplicado, también fuimos informados por la fuente del dataset de que los datos vienen sin formato, es decir, sin identificar como números u otros tipos, por ello, reparamos el `DataFrame`  con la función `repair_data` definida a continuación.





```python
def repair_data(data):
    if "Unnamed: 0" in data.columns:
        data = data.drop("Unnamed: 0", axis = 1)
    for col in data.columns[:-1]:
        data[col] = pd.to_numeric(data[col],
                                  downcast = "float",
                                  errors = "coerce")
    return data

STARS_DATA = repair_data(RAW_DATA.copy())
STARS_DATA.describe()
```

## Filtrado de los datos

Podemos notar que no todos los datos contienen la misma cantidad de información, como no pasa de las 100 filas el máximo, sabemos que en relación al tamaño actual del conjunto de datos es una cantidad menor, por lo que optamos por comenzar a descartar información. Cabe mencionar que la medición de la columna `Plx` corresponde a la distancia entre la Tierra y esa estrella mediante la técnica de paralaje estelar, (*Parallax*) y a su vez, el atributo `e_Plx` denota el porcentaje de error de esta medición, para obtener un 98% de fiabilidad de estos datos, eliminaremos las filas que no cumplan ese margen de error, y también las que tengan una medición `Plx` igual a 0, pues estas estrellas estarán tan cerca que perderá significado analizarlas.

Adicionalmente, se eliminarán mediciones de estrellas cuya medición sea imprecisa a causa de particularidades. En la clasificación espectral (`SpType`) existe una nomenclatura para representar todo tipo de peculiaridades, desde la detección de emisión de algún tipo particular de radiación o elemento, pasando por características de sus líneas de emisión hasta incluso destacar variaciones con respecto a su medición. Son estos últimos en los que nos fijaremos para discernir posibles bajas de precisión. En la nomenclatura mencionada, los siguientes caracteres podrían significar problemas con su posterior clasificación dentro del diagrama HR:

* Con un `:` se indica que el tipo espectral (referente a su color/temperatura) es incierto.
* Con un `/` se indica que la detección podría se o bien la indicada a la izquierda de este separador, o la indicada a la derecha.
* Con un `-` similar a lo que ocurre con el `/` separa entre una y otra medición, solo que esta indica que la medición debiera ubicarse entre una o la otra. Similarmente se utiliza `.` para indicar fracciones en dígitos, pero esto no representa problema, ya que los dígitos indican un sub tipo espectral del prefijo a este.
* Con un `+` se indica que la medición en realidad representa un cúmulo de estrellas, lo que significa que la medición contempla más de una estrella para su clasificación.

```python
def filter_data(data):
    raw = data.shape[0]
    data = data.dropna()
    nan_not_dropped = data.shape[0]
    nan_drop_percent = (raw - nan_not_dropped) * 100 / raw
    print(f"The {nan_drop_percent:.2f}% of the rows were dropped, for being incomplete.")
    data = data[data["e_Plx"] < (100 - 98)]
    data = data[data["Plx"] != 0]
    error_not_dropped = data.shape[0] + (raw - nan_not_dropped)
    error_drop_percent = (raw - error_not_dropped) * 100 / raw
    print(f"The {error_drop_percent:.2f}% of the rows were dropped, to prevent errors.")
    avoid = [":", "/", "-", "+"]
    data = data[list(map(lambda x: not any(list(map(lambda i: i in x, avoid))), data["SpType"]))]
    accuracy_dropped = data.shape[0] + (raw - error_not_dropped)
    accuracy_drop_percent = (raw - accuracy_dropped) * 100 / raw
    print(f"The {accuracy_drop_percent:.2f}% of the rows were dropped, due to peculiarities.")
    return data

STARS_DATA = filter_data(STARS_DATA)
```

# Clasificación de los datos

Como contamos con información reducida acerca de las estrellas, pero si tenemos la clasificación espectral `SpType` de cada una, con lo que podemos generar una clasificación más reducida sobre las estrellas. Tal como se mencionó en principio, resulta interesante comenzar por la clasificación dada por el diagrama HR, es por esto que empleando expresiones regulares podemos detectar de forma medianamente aceptable a que clasificación del diagrama corresponde cada estrella.

Ahora bien, tal como hemos podido notar hasta ahora en el trabajo, el estado del conjunto de datos deja un poco que desear, y la clasificación espectral no es la excepción, puesto que hay una enorme variación de las fuentes utilizadas para la clasificación, incluso empleando algunas bajo documentos del 75', variantes de medición entre otros que podemos observar al revisar el repositorio de Harvard referenciado al final de este trabajo, o notando la existencia del sufijo `SB` o `var` al final de la clasificación que implican que la medición se hizo mediante un Espectroscopio Binario, o que se observó variaciones en la medición respectivamente.

Bajo todos estos inconvenientes que podrían reducir la precisión, la siguiente función obtiene la clasificación bajo el diagrama HR para algunas estrellas que presenten en particular la clase de luminosidad en su clasificación previa de clase espectral, tomando en consideración también su tipo espectral para algunos rangos de temperatura aproximados que nos permitan ajustar un poco mejor los resultados a los que se espera.

```python
def classify_data(data):
    W = "(D[A])|([A].*VII)"
    G = "[GKM][^V]+I{2}|F[^VI]+I{2}[^VI]+"
    M = "I{2}|(sd)|([OBAFGKM][^V]+V)"
    H = "[OBAFGKM][^V]I[^VI]+"
    result, classes = [], {"W":W, "G":G, "H":H, "M":M}
    for st in data["SpType"]:
        switch = True
        for c in classes:
            if re.search(classes[c], st + "_"):
                result.append(c)
                switch = False
                break
        if switch:
            result.append("S")
    return result

STARS_DATA["Class"] = classify_data(STARS_DATA)
```

Bajo el mismo propósito de luego poder clasificar las estrellas mediante modelos predictivos en la clasificación del diagrama HR, nos interesa obtener algo más de información que la que tenemos actualmente. Calcularemos la magnitud absoluta `Amag`, puesto que solo contamos con la magnitud visualmente aparente `Vmag`, para ello usamos la fórmula:

$$M = m + 5(log_{10}p + 1)$$

Adicionalmente, podemos obtener una aproximación de la temperatura en grados Kelvin de la estrella a partir del color de la estrella, medición que se encuentra descrita en el atributo `B-V` que obtiene el índice del color en base a esa escala que emplea sístema fotométrico UBV, este sistema, en resumen, son filtros que observan la luz como una onda y separan la luz en 3 longitudes de onda, notación que se llevó a cabo por especificaciones de un modelo de telescopio usado en ese entonces, pero que se conservan al día de hoy. La aproximación de temperatura se calcula de la siguiente forma:

$$T = 4600\left(\frac{1}{0.92B-V + 1.6} + \frac{1}{1.5B-V + 0.62}\right)$$

Luego, podemos calcular la luminosidad relativa a la luminosidad punto cero a partir de la magnitud absoluta, para más información sobre este cálculo y la gran discusión que hay en el área sobre considerar o no la magnitud del Sol como un referente para este cálculo, se puede visitar la página de la definición de luminosidad adjunta en el apartado de referencias. En resumen, habitualmente se utilizaba la magnitud bolométrica envés de la absoluta, bajo una corrección con la magnitud bolométrica del Sol, pero a partir de algunos análisis hechos por expertos en el área, el error entre ambos cálculos se ha considerado despreciable para el autor de esta aproximación. A raíz de la simplificación que significa, y de que no contamos con mediciones de la magnitud bolométrica en este `DataFrame`, optamos por calcular la luminosidad bajo la relación:

$$\frac{L_*}{L_0} = 10^{-0.4M}$$

Finalmente, ya que tenemos la luminosidad y la temperatura, podemos calcular el radio aproximado de la estrella, empleando la siguiente ecuación que aparece de un despeje en la ley de radiación Stefan-Boltzmann:

$$R = \left(\frac{L_*}{4πσT^4}\right)^{0.5}$$

Cabe mencionar que existen algunas diferencias en base a las unidades de medida, estas se tratarán en el código, igualmente para el cálculo del radio se utilizará la constante de la luminosidad punto cero para obtener la luminosidad total. Calculamos la luminosidad relativa, puesto que sin escalar toma números demasiado grandes que pueden afectar negativamente por estar en otro nivel de magnitud en comparación al resto de datos.

```python
def increment_data(data):
    sigma, zp = 5.67e-8, 3.0128e28 # sigma and zero point constants
    data["Amag"] = np.log10(abs(data["Plx"] / 1000)) # milliarcseconds to arcseconds
    data["Amag"] = data["Vmag"] + 5 * (data["Amag"] + 1)
    data["Temp"] = 1 / (0.92 * data['B-V'] + 1.6)
    data["Temp"] = 4600 * (data["Temp"] + 1 / (1.5 * data['B-V'] + 0.62))
    data["Rlum"] = (10 ** (data["Amag"] * -0.4))
    c = 4 * np.pi * sigma
    data["Rad"] = ((data["Rlum"] * zp) / (c * (data["Temp"] ** 4))) ** 0.5
    return data

STARS_DATA = increment_data(STARS_DATA)
```

Por último, la función `resample_data` se encargará de quitar la clase anterior `SpType` y separar el conjunto entre las estrellas cuya clasificación dentro del diagrama HR haya sido detectada de las que no, bajo el parámetro binario `get_both` se podrá recuperar las que no cuenten con clasificación, lo que podría ser útil para proósitos de prueba de la robustes de la solución a la que lleguemos.

```python
def resample_data(data, get_both = False):
    every_star = data.shape[0]
    data = data.drop("SpType", axis = 1)
    classified_data = data[data["Class"] != "S"]
    unclassified_data = data[data["Class"] == "S"]
    interesting_stars = data.shape[0]
    droped_rows = every_star - interesting_stars
    print(f"A total of {droped_rows} rows are classified, the rest lack of a class.")
    if get_both:
        aux = unclassified_data.reset_index(drop = True)
        return classified_data.reset_index(drop = True), aux
    return classified_data.reset_index(drop = True)

STARS_DATA = resample_data(STARS_DATA)
```

Decidimos agregar un nuevo set de datos, pudimos tomar este camino en base a que todo lo anterior lo definimos en forma de función, por lo que, la exploración inicial de los datos iba a funcionar ya sea en este o en otro dataframe que se basara en HR y que tuviera indice espectral. La razón principal de esta decisión es que el csv inicial tenía muy pocos datos y ninguna enana blanca, por lo que quisimos tener acceso a una base que tuviera más información.

```python
NEW_DATA = pd.read_csv(REPO + "more_stars.csv")
MORE_STARS = repair_data(NEW_DATA)
MORE_STARS = filter_data(MORE_STARS)
MORE_STARS = increment_data(MORE_STARS)
MORE_STARS["Class"] = classify_data(MORE_STARS)
MORE_STARS, BLIND_STARS = resample_data(MORE_STARS, get_both = True)
MORE_STARS.tail()
```

Le aplicamos al nuevo dataset las funciones necesarias para limpieza, filtrado y exploración. Ahora, realizaremos un gráfico que nos muestre la posición relativa de las estrellas de cada dataset respecto al diagrama HR.

```python
def hr_diagram(data):
    by_classes = [
        data.loc[data["Class"] == "M"], data.loc[data["Class"] == "G"],
        data.loc[data["Class"] == "H"], data.loc[data["Class"] == "W"]
    ]
    labels = ["Main Sequence stars", "Giant stars", "Super/Hyper giants", "White dwarfs"]
    plt.title("Hertzsprung–Russell diagram")
    for d, l in zip(by_classes, labels):
        plt.plot(np.log10(d["Temp"]), np.log(d["Rlum"]), ls = "", marker = ".", label = l)
    plt.xlabel("Temperature in Log K")
    plt.xlim(np.log10(27000), np.log10(1500))
    plt.ylabel("Luminosity in Log L0 watts")
    plt.ylim(-14, 14)
    plt.legend(loc = 3)

plt.figure(figsize = (16, 10))
plt.subplots_adjust(wspace = 0.2)
plt.subplot(1, 2, 1)
hr_diagram(STARS_DATA)
plt.subplot(1, 2, 2)
hr_diagram(MORE_STARS)
```

Como se pronosticaba en el apartado de clasificación, nuestra clasificación manual tiene variados problemas, los que esperamos poder resolver mediante modelos predictivos con aprendizaje supervisado y no supervisado.

Además, nos gustaría hacer una observación que es algo grave al respecto con nuestro objetivo original, y eso es que el conjunto contaba con una cantidad muy reducida de estrellas enanas blancas, lo cual tiene sentido dado que son naturalmente raras, pero resulta muy inconveniente para nuestro objetivo.

Por todo lo mencionado anteriormente, optamos por agregar una métrica adicional a las convencionales, que será la observación del gráfico con el diagrama HR usando la predicción del modelo como nueva clase, cosa de, idealmente, poder ir continuamente mejorando la clasificación con la que hemos comenzado.

```python
def predict_hr(X, y, name):
    columns = ["Vmag", "Plx", "e_Plx", "B-V", "Amag", "Temp", "Rlum", "Rad"]
    classified = pd.DataFrame(X, columns = columns)
    try:
        classified["Class"] = ["MHGW"[i] for i in y]
    except:
        classified["Class"] = [str(i) for i in y]
    hr_diagram(classified)
    plt.title(f"Hertzsprung–Russell diagram for {name}.")
```

Para facilitar la observación de las clasificaciones generadas por las distintas estrategias empleadas a continuación, se define la última función `predict_hr`, que genera un diagrama HR a partir de la predicción obtenida por el método que se haya utilizado.

## Aprendizaje No Supervisado (K-Means)

Decidimos utilizar K-means (o visto de manera general un método particional), ya que, en función de los ordenes numéricos y además la cantidad de datos que contiene un dataset de tipo astronómico estimamos que usar clustering de tipo jerárquico sería muy complejo de análizar (probablemente los dendogramas serían sólo ruido), sin contar que serían increíblemente lentos y de mucho costo computacional.

No requeriremos de usar métricas para conocer el K ideal, ya que, a nivel humano ya conocemos la cantidad de clusteres que necesitamos realizar, por lo que esta inspección se basará netamente en ver que tan bien puede un algoritmo no supervisado trabajar con la información de las estrellas.

```python
def evaluate_clustering(estimator, X, y, name = "KMeans"):
    nX = skp.scale(X.copy())
    ini = time.time()
    est = estimator.fit(nX)
    end = time.time()
    pred, y = est.labels_, np.array(y)
    predict_hr(X, pred, name)
    metrics = {
        "Completeness": [skm.completeness_score(y, pred)],
        "Homogeneity": [skm.homogeneity_score(y, pred)],
        "V-Measure": [skm.v_measure_score(y, pred)],
        "Silhouette": [skm.silhouette_score(nX, pred, random_state = SEED)],
        "Inertia": [est.inertia_],
        "Convergence Time": [end - ini]
    }
    return pd.DataFrame(metrics, index = [name])

n = np.unique(STARS_DATA["Class"]).shape[0]
km = skc.KMeans(n_clusters = n, random_state = SEED)
X = STARS_DATA.loc[:, STARS_DATA.columns != "Class"]
evaluate_clustering(km, X, STARS_DATA["Class"])
```

Tal y como teníamos en mente, el cluster funcionó bastante mal para la clasificación inicial, si bien es cierto que `silhouette` tiró un muy buen puntaje, es fácil notar que la densidad que tiene los grupos, especialmente el azul es muy alta, lo cual influye directamente en estos resultados. No obstante podemos ver que `completeness` da un valor considerablemente bajo. No es difícil notar que esto se debe a que muchos puntos que originalmente eran naranjas y verdes quedaron dentro de la nube azul, estimábamos que algo así pasaría por la forma de los grupos. No obstante por la cantidad de datos y por la forma (no habríamos sabido a priori que `linkeage` utilizar) decidimos descartar jerárquico, así como también utilizar centros en base a observación, esto debido a que, como se puede ver, el gráfico es de temperatura contra luminosidad, lo cual realmente no nos serviría para explorativamente obtener centros mejores, ya que el cluster de por sí utilizaría todas las variables y no sólo las 2 que contempla el gráfico.

### Aprendizaje supervisado (SVM y Naive Bayes)

Decidimos utilizar SVM multiclase para ver que tal hacía estas separaciones de clases. Sabemos que SVM es un algoritmo muy potente y queremos comparar resultados respecto a lo que serían capaces de hacer las redes neuronales. Naive Bayes, como vimos en cátedra, suele funcionar bien de manera natural con multiclase, así que lo usaremos para realizar comparaciones. Por la naturaleza probabilística y en base a la función masa de probabilidad esperamos este pueda hacer una buena tarea de clasificación.

```python
def get_predictions(classifiers, names, method = "cross_validate"):
    predictions = {}
    for clf, name in zip(classifiers, names):
        if method == "cross_validate":
            predictions[name] = sms.cross_val_predict(clf, TRAIN_X, TRAIN_Y, cv = 4)
        elif method == "local_fit":
            predictions[name] = (clf.fit(TRAIN_X, TRAIN_Y)).predict(TEST_X)
        else:
            raise f"Parameter 'method' had a not defined value: {method}."
    return predictions

def metric_comparison(predictions, names, y):
    simplefilter("ignore", category = UndefinedMetricWarning)
    data = {
        "Precision" : [], "Recall" : [], "F1" : [], "Accuracy" : []
    }
    for p in predictions:
        report = skm.classification_report(p, y, output_dict = True)
        data["Precision"].append(report["macro avg"]["precision"])     
        data["Recall"].append(report["macro avg"]["recall"])
        data["F1"].append(report["macro avg"]["f1-score"])
        data["Accuracy"].append(report["accuracy"])
    return pd.DataFrame(data, names)
```

Creamos funciones para predecir que tal están funcionando los clasificadores, se utilizará validación cruazada de modo de obtener resultados de alta calidad.

Además, generamos las funciones suficientes para poder comparar las métricas clásicas. Ante esto procedemos a realizar los experimentos correspondientes.

```python
def metric_model(prediction, Y, model_name = ""):
    conf_matrix = skm.confusion_matrix(prediction, Y)
    names = {
        "xticklabels" : ["G", "H", "M"],
        "yticklabels" : ["G", "H", "M"]
    }
    sns.heatmap(conf_matrix, cbar = False, annot = True, **names)
    opt_space = ' ' if model_name != "" else ''
    plt.title(f"Confusion Matrix of{opt_space}{model_name}.")
    plt.xlabel("Actual value")
    plt.ylabel("Prediction")
    return conf_matrix

mcc = {
    "RBF SVM" : sks.SVC(decision_function_shape='ovo', random_state = SEED),
    "Naive Bayes" : skn.GaussianNB()
}
cross_preds = get_predictions(mcc.values(), mcc.keys())

plt.figure(figsize = (20, 4))
plt.subplots_adjust(wspace = 0.33)
for index, name in enumerate(cross_preds):
    plt.subplot(1, 4, index + 1)
    null = metric_model(cross_preds[name], TRAIN_Y, name)
```

Después de diversos experimentos podemos ver que ni SVM ni Naive Bayes han funcionado de muy buena manera, si bien es cierto que al ojo parecieran que funcionan mejor que los clustering, el caso es que la variación es leve y aún así, en efecto, funcionan mal. principalmente por la gran cantidad de casos donde se dio la clasificación M siendo que eran G. ### Aprendizaje supervisado (Random Forest)

El clasificador de aprendizaje supervisado que mencionaremos a continuación funciona bien para multiclase que es algo que para nuestro trabajo nos interesa y al no ser visto en el curso consideramos bueno explicarlo detalladamente para entender su funcionamiento.

Random Forest es un modelo de aprendizaje supervisado para clasificación (aunque también puede usarse para problemas de regresión).  Es un tipo de Ensamble en Machine Learning en donde combinaremos diversos árboles de decisión en donde se entrenará cada árbol de decisión en una muestra diferente de las observaciones.

Algunos de los beneficios que tiene random forest son por ejemplo que en comparación con los árboles de decisión individuales los cuales tienden a sobre ajustarse (overfit) a los datos de entrenamiento, el bosque aleatorio puede mitigar ese problema al promediar los resultados de predicción de diferentes árboles lo que da al algoritmo de bosques aleatorios una mayor precisión predictiva que un solo árbol de decisión. También funciona bien para múltiples muestras. 


image.png

Como vemos en la imagen random forest funciona de la siguiente manera:
* Seleccionamos k features (columnas) de las m totales (siendo k menor a m) y creamos un árbol de decisión con esas k características.
* Creamos n árboles variando siempre la cantidad de k features y también podríamos variar la cantidad de muestras que pasamos a esos árboles (esto es conocido como “bootstrap sample”)
* Tomamos cada uno de los n árboles y le pedimos que hagan una misma clasificación. Guardamos el resultado de cada árbol obteniendo n salidas.
* Calculamos los votos obtenidos para cada “clase” seleccionada y consideraremos a la más votada como la clasificación final de nuestro “bosque”.

Ya presentado este nuevo modelo predictivo, pasamos a realizar pruebas sobre el conjunto inicial. En principio, la operabilidad del clasificador de bosque aleatorio es equivalente a la de cualquier estimador de la librería `sklearn`, por lo que las pruebas a continuación son bastante estándar, obviando por supuesto nuestra métrica del diagrama HR.

### Aprendizaje supervisado (Redes Neuronales)

Como ya dimos a entender, también utilizaremos redes neuronales para analizar nuestro trabajo. Para este caso utilizaremos solo una capa oculta. Ya que como pudimos observar en el trabajo anterior, si puede existir una mejora de precisión, pero a un costo computacional demasiado grande como para considerarlo, así que en pro de terminar antes con las pruebas y de simplificar el análisis que hasta ahora ya ha sido extenso, optamos por mantener las capas ocultas limitadas a una.

En astronomía las redes neuronales son muy utilizadas dado que dan muy buenos resultados cuando se esta trabajando con multiclase, algo que no necesariamente podemos ver cuando trabajamos con SVM. La posibilidad de generar pertubaciones no lineales estimamos nos sea de especial utilidad, probaremos para distintas funciones de activación de modo que podamos hacer un análisis bastante completo.

Por esta razón, comenzamos por probar la precisión del modelo. A continuación el código revisa la accuracy obtenida por el modelo en un rango variable, luego de algunas pruebas determinamos que el mejor rango de prueba fue de 30 a 90 nodos. Además, estaremos utilizando el dataset `MORE_STARS` en principio, ya que como pudimos observar en trabajos anteriores, las redes neuronales tienden a trabajar mejor con volúmenes grandes de datos.


```python
def pcopy(perceptron):
    kwargs = perceptron.get_params()
    new_perceptron = snn.MLPClassifier()
    return new_perceptron.set_params(**kwargs)

def perceptrons_comparison(models, attr, labels, form = lambda i: [i]):
    size = len(models)
    y_model = [list() for i in range(size)]
    best, best_model = [0 for i in range(size)], [None for i in range(size)]
    for i in range(30, 91):
        for j in range(size):
            model = models[j].set_params(**{attr : form(i)})
            s_model = model.fit(TRAIN_X, TRAIN_Y).score(TEST_X, TEST_Y)
            y_model[j].append(s_model)
            if (s_model > best[j]):
                best[j], best_model[j] = s_model, pcopy(model)
    plt.title(f"Accuracy comparison for {labels[0]}.")
    x = [i for i in range(30, 91)]
    for i, y in enumerate(y_model):
        plt.plot(x, y, label = labels[i + 1])
    plt.xlabel(attr)
    plt.ylabel("accuracy_obtained")
    plt.legend()
    return best_model

set_dataset(MORE_STARS[MORE_STARS["Class"] != "W"])
perceptrons = [
    snn.MLPClassifier(max_iter = 1000, activation = "relu", early_stopping = True, random_state = SEED),
    snn.MLPClassifier(max_iter = 1000, activation = "identity", early_stopping = True, random_state = SEED)
]
labels = ["2 Neural Networks by activation function", "ReLu activation", "Linear activation"]
one_layer_bests = perceptrons_comparison(perceptrons, "hidden_layer_sizes", labels)
```

Después de realizar diversos experimentos, optamos por quedarnos con Relu e identidad, ya que sigmoide y tanh no nos dieron valores muy trabajables. En el código a continuación podremos ver la cantidad de nodos necesarios para obtener la accuracy máxima obtenida por cada función de activación. 

```python
for p in one_layer_bests:
    params = p.get_params()
    print(f"For activation with the {params['activation']} function,", end = " ")
    print(f"it were needed {params['hidden_layer_sizes'][0]} nodes to max the accuracy.")
```

Y bajo la misma lógica de los estimadores anteriores, probamos las métricas de certeza para los mejores clasificadores obtenidos, junto con la métrica del diagrama HR.

```python
one_layer_names = ["ReLu 77", "Linear 76"]
nn_preds = get_predictions(one_layer_bests, one_layer_names, method = "local_fit")
plt.figure(figsize = (12, 6))
plt.subplots_adjust(wspace = 0.26)
for i, n in enumerate(one_layer_names):
    plt.subplot(1, 2, i + 1)
    predict_hr(TEST_X, nn_preds[n], n)
metric_comparison(nn_preds.values(), one_layer_names, TEST_Y)
```

Podemos ver que, en contrario a lo que creíamos, las redes neuronales no funcionaron especialmente bien, estimamos que tiene relación directa con la cantidad de estrellas super/hiper gigantes, ya que hace un corte con mucha similaridad entre Main Squence y Giant respecto a lo que el HR realmente hace, sólo que de manera "poco natural" a lo que podemos entender a nivel humano, la curva que generaron las redes, tanto con ReLu como con Linear es una separación binaria poco suave.
