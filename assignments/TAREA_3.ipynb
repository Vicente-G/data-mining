## **1.2** Realice una breve descripción de los siguientes clasificadores, considerando su uso en clasificación multiclase.
- Suport vector Machine (SVM): Es un modelo de representación de puntos en el espacio, este mismo se basa en métricas de distancia, lo que busca principalmente es separar de la manera más amplia posible los puntos para que, en función de su posición relativa respecto al hiperplano que separa, dar una clasificación a los objetos. One-Vs-All es un algoritmo que funciona con entradas binarias y se compara en iteraciones que contempla a todas las posibles clases, mientras que en One-Vs-One se hace comparaciones de cada clase contra una sóla a la vez.

Específicamente para multiclase lo que hace es realizar la comparación probabilística para cada categoría necesaria. Por su interpretación matemática, no hay diferencias considerables entre usar Naive Bayes para un problema de una clase o multiclase.

 - Pesos y Bias: Los pesos son alteraciones de la señal recibida por la red, hacen que la RNA sea capaz de aplicar soluciones en base a coeficientes, estas pueden ser modificadas en el entrenamiento para obtener resultados más precisos.
```python
def img_to_df(url, categories):
    X, y = [], []
    for animal in categories:
        print(f"Loading {animal} category...")
        path = os.path.join(url, animal)
        file_list = requests.get(path + ".txt").text.split("\r\n")[:-1]
        for filename in file_list:
            img_array = imread(os.path.join(path, filename))
            img_resized = resize(img_array, (150, 150, 3))
            featured_data, _ = hog(
                img_resized,
                orientations = 8,
                pixels_per_cell = (16, 16),
                cells_per_block = (1, 1),
                visualize = True
            )
            X.append(featured_data)
            y.append(categories.index(animal))
        print(f"Category {animal} loaded successfully!")
    df = pd.DataFrame(np.array(X))
    df['Target'] = np.array(y)
    return df

animals = ["Beagle", "Bengal", "Birman", "Bombay"]
url = "https://raw.githubusercontent.com/Vicente-G/data-mining/main/"
df = img_to_df(url + "animals/", animals)
df.to_csv("animals.csv") # downloaded here and uploaded on GitHub
```

# Clasificadores multiclase
A continuación 2 métodos auxiliares para simplificar la forma en que se obtienen clasificadores multiclase y las predicciones de un clasificador cualquiera.

```python
def get_multiclass(base_classifiers, names):
    classifiers = {}
    for clf, name in zip(base_classifiers, names):
        classifiers[name + " with OVR"] = smc.OneVsRestClassifier(clf)
        classifiers[name + " with OVO"] = smc.OneVsOneClassifier(clf)
    return classifiers

def get_predictions(classifiers, names, method = "cross_validate"):
    predictions = {}
    for clf, name in zip(classifiers, names):
        if method == "cross_validate":
            predictions[name] = sms.cross_val_predict(clf, TRAIN_X, TRAIN_Y, cv = 4)
        elif method == "local_fit":
            predictions[name] = (clf.fit(TRAIN_X, TRAIN_Y)).predict(TEST_X)
        else:
            raise f"Parameter 'method' had a not defined value: {method}."
    return predictions
```

Bajo la consideración de los métodos definidos anteriormente, se define una función para obtener y graficar la matriz de confusión de un clasificador en base a su predicción. En este caso, se observa el comportamiento multiclase de los clasificadores Super Vector Machine (SVM) y Naive Bayes (NB), bajo las estrategias multiclase de One Versus Rest (OVR) y One Versus One, (OVO) mediante la observación de matrices de confusión obtenidas con la predicción promedio de aplicar validación cruzada sobre los clasificadores.

```python
def metric_model(prediction, Y, model_name = ""):
    conf_matrix = skm.confusion_matrix(prediction, Y)
    names = {
        "xticklabels" : np.unique(prediction),
        "yticklabels" : np.unique(prediction)
    }
    sns.heatmap(conf_matrix, cbar = False, annot = True, **names)
    opt_space = ' ' if model_name != "" else ''
    plt.title(f"Confusion Matrix of{opt_space}{model_name}.")
    plt.xlabel("Actual value")
    plt.ylabel("Prediction")
    return conf_matrix

svm = sks.LinearSVC(dual = False, random_state = SEED)
gau = skn.GaussianNB()
mcc = get_multiclass([svm, gau], ["SVM", "NB"])
cross_preds = get_predictions(mcc.values(), mcc.keys())

plt.figure(figsize = (20, 4))
plt.subplots_adjust(wspace = 0.33)
for index, name in enumerate(cross_preds):
    plt.subplot(1, 4, index + 1)
    null = metric_model(cross_preds[name], TRAIN_Y, name)
```

Para un análisis más claro de lo que ocurre con estos clasificadores, podemos hacer un recuento rápido de las métricas a las que estamos habituados, utilizando los resultados anteriores.

```python
def metric_comparison(predictions, names, Y):
    data = {"Accuracy" : [], "Precision" : [], "Recall" : [], "F1" : []}
    for prediction in predictions:
        conf_matrix = skm.confusion_matrix(prediction, Y)
        precision, recall = [], []
        for i in range(len(conf_matrix)):
            total_p, total_r = 0, 0
            for j in range(len(conf_matrix)):
                total_p += conf_matrix[j, i]
                total_r += conf_matrix[i, j]
            precision.append(conf_matrix[i, i] / total_p)
            recall.append(conf_matrix[i, i] / total_r)
        data["Accuracy"].append(sum(prediction == Y) / len(Y))
        data["Precision"].append(np.average(precision))
        data["Recall"].append(np.average(recall))
        precision, recall = np.average(precision), np.average(recall)
        data["F1"].append(2 * (precision * recall) / (precision + recall))
    return pd.DataFrame(data, names)

metric_comparison(cross_preds.values(), mcc.keys(), TRAIN_Y)
```

A modo general, ninguno de los clasificadores tiene una buena tasa de aciertos, sin importar mayormente la estrategia utilizada. Ahora bien, observamos a continuación como se desempeñan estos clasificadores con el conjunto de prueba.

```python
local_preds = get_predictions(mcc.values(), mcc.keys(), method = "local_fit")
metric_comparison(local_preds.values(), mcc.keys(), TEST_Y)
```

Como se puede observar, no existe una mejora con respecto al panorama general, más bien todo lo contrario. Por lo demás, los resultados obtenidos nos permiten concluir empíricamente que para este conjunto de datos, es mejor utilizar la estrategia OVO de multiclase, ya que existe una mejora entre usar una opción o la otra.

Esto por supuesto que nos hace sentido, dada la naturaleza del conjunto de datos, que presenta 648 atributos por instancia, lo que hace que la combinatoria de estos permita discernir más facilmente una de otra cuando se enfrentan una a una. No así si se considera que el resto de clases y sus múltiples combinatorias recaen en una sola clase. Además, el dataset se encuentra balanceado con 200 instancias de cada clase, por lo que se genera un desbalance al emplear la estrategia de OVR, lo que puede recaer en la disminución de precisión del clasificador.

# Perceptrones multicapa
Para analizar los perceptrones de solo una capa oculta de una sola vez, utilizamos un razonamiento similar al de trabajos anteriores, generando un gráfico que modificara el número de nodos de la capa para múltiples modelos a la vez y graficara la `Accuracy` de cada uno en un gráfico.

Esto nos permitiría observar de forma sencilla el comportamiento para determinados numeros de nodos para la capa del perceptrón, bajo distintas funciones de activación. De este proceso se encarga la función `perceptrons_comparison` definida más abajo. Además, esta retornará en una lista los mejores perceptrones para cada función.

Dado que la combinatoria posible para los parámetros de los perceptrones es inmensa, optamos por apegarnos solo a un rango de nodos (70 a 150) para la primera capa, por lo que se podría esperar que exista un mejor resultado fuera de este rango. Además, se fijará el máximo de generaciones en 1000, para evitar que los algoritmos tomen demasiado tiempo, dentro de lo posible.

```python
def pcopy(perceptron):
    kwargs = perceptron.get_params()
    new_perceptron = snn.MLPClassifier()
    return new_perceptron.set_params(**kwargs)

def perceptrons_comparison(models, attr, labels, form = lambda i: [i]):
    size = len(models)
    y_model = [list() for i in range(size)]
    best, best_model = [0 for i in range(size)], [None for i in range(size)]
    for i in range(70, 151):
        for j in range(size):
            model = models[j].set_params(**{attr : form(i)})
            s_model = model.fit(TRAIN_X, TRAIN_Y).score(TEST_X, TEST_Y)
            y_model[j].append(s_model)
            if (s_model > best[j]):
                best[j], best_model[j] = s_model, pcopy(model)
    plt.title(f"Accuracy comparison for {labels[0]}.")
    x = [i for i in range(70, 151)]
    for i, y in enumerate(y_model):
        plt.plot(x, y, label = labels[i + 1])
    plt.xlabel(attr)
    plt.ylabel("accuracy_obtained")
    plt.legend()
    return best_model

perceptrons = [
    snn.MLPClassifier(max_iter = 1000, activation = "relu", early_stopping = True, random_state = SEED),
    snn.MLPClassifier(max_iter = 1000, activation = "logistic", early_stopping = True, random_state = SEED),
    snn.MLPClassifier(max_iter = 1000, activation = "tanh", early_stopping = True, random_state = SEED)
]
labels = ["3 perceptrons by activation", "ReLu activation", "Sigmoid activation", "tanh activation"]
one_layer_bests = perceptrons_comparison(perceptrons, "hidden_layer_sizes", labels)
```

Como se puede observar, ninguna de las funciones de activación prevalece de forma continua sobre las otras, de hecho, es difícil obtener cuál fue el máximo peak en este gráfico, pero mediante el próximo código aclararemos el que clasificadores maximizaron la accuracy.

El algoritmo es sencillo, en la variable `distributions` se generan aleatoriamente 2 listas, de 2 y 3 números para representar las múltiples capas, como no están asociados a una semilla, y por el método en sí, presentan valores aleatorios distintos cada vez. Luego, con la función `RandomizedSearchCV` se obtiene la solución que maximiza la accuracy del perceptrón bajo las combinaciones de la i-ésima iteración. El algoritmo solo se detendrá cuando el perceptrón obtenido, tenga un mejor desempeño que el mejor clasificador de la etapa anterior.

El algoritmo podría tardar muchísimas iteraciones en terminar, o para efectos prácticos, podría no terminar, por eso se encuentra limitado a una sola iteración, pues en el código siguiente se utilizan parámetros conocidos que sobrepasan la accuracy anterior.

```python
simplefilter("ignore", category = UserWarning)
new_acc = 0
while (best_acc > new_acc):
    distributions = {'hidden_layer_sizes': [(spr.rvs(70,150,1), spr.rvs(40,70,1), spr.rvs(10,40,1),), (spr.rvs(70,150,1), spr.rvs(10,70,1),)],'activation': ['tanh', 'relu', 'logistic']}
    clf = snn.MLPClassifier(max_iter = 1000, early_stopping = True, random_state = SEED)
    search = sms.RandomizedSearchCV(clf, distributions, random_state = SEED)
    search.fit(TRAIN_X, TRAIN_Y)
    new_acc = search.score(TEST_X, TEST_Y)
    break # comment this line at your own risk
params = search.best_params_
print(f"The accuracy found in this iteration was: {new_acc * 100}%")
```

A continuación se entrena una copia del clasificador que mejores resultados nos ha dado en las pruebas, bajo los parámetros que se han considerado a lo largo de este trabajo, es decir, el número limitado de nodos y capas, la semilla utilizada, el dataset elegido, el tope de iteraciones y más.

```python
params = {"hidden_layer_sizes" : (133, 55, 26,), "activation" : "relu"}
best_model = snn.MLPClassifier(max_iter = 1000, early_stopping = True, random_state = SEED, **params)
print(f"Best accuracy found: {best_model.fit(TRAIN_X, TRAIN_Y).score(TEST_X, TEST_Y) * 100}%")
```

```python
def loss_computation(models, labels):
    plt.title("Perceptron loss comparison.")
    for model, name in zip(models, labels):
        clf = pcopy(model).fit(TRAIN_X, TRAIN_Y)
        y, iters = clf.loss_curve_, clf.n_iter_
        plt.plot([x + 1 for x in range(iters)], y, label = name)
    plt.xlabel("Iteration number")
    plt.ylabel("Loss amount")
    plt.legend()
    plt.show()

loss_computation(one_layer_bests + [best_model], one_layer_names + ["Multilayer"])
```

# Selección de características
Continuando sobre la idea general de este trabajo de una mejora continua en la clasificación para el problema de multiclase con el que hemos comenzado, visitaremos otra de las perspectivas que se emplean para buscar mejoras en la clasificación sobre un dataset. Hablamos de la selección de características, en particular, comenzaremos por utilizar la función `SelectKBest`, junto al criterio de puntuación $chi^2$ para reducir la cantidad de atributos que tiene el conjunto de datos.

Para ello, le pasaremos a la función el conjunto completo y luego se dividirá en conjunto de entrenamiento y prueba bajo los mismos parámetros utilizados anteriormente. Finalmente, se calculará la accuracy que tenga el mejor clasificador hasta ahora sobre el nuevo conjunto de datos con menos atributos.

```python
def k_accuracy(clf, X, y, k_array = [10, 200, 400, 500, 600]):
    plt.title("Accuracy obtained with best features.")
    accuracy = []
    for i in k_array:
        new_X = skf.SelectKBest(skf.chi2, k = i).fit_transform(X, y)
        tX, TX, ty, Ty = sms.train_test_split(new_X, y, test_size = 0.25, random_state = SEED)
        clf.fit(tX, ty)
        accuracy.append(clf.score(TX, Ty))
    plt.plot(k_array, accuracy)
    plt.xlabel("feature_quantity")
    plt.ylabel("accuracy")
    plt.show()

k_accuracy(best_model, ANIMALS_X, ANIMALS_Y)
```

Por desgracia, podemos observar que con esta estrategia, al menos con los valores sugeridos, no existe una mejora en comparación con la accuracy obtenida anteriormente. No contentos con dicho resultado, empleamos otra estrategia de la misma disciplina, esta vez, con la función `SequentialFeatureSelector`.

```python

def time_it(process):
    start = time.time()
    result = process()
    done = time.time() - start
    return result, done

def fs_accuracy(clf, X, y, f_select = [3, 5, 7, 10]):
    plt.figure(figsize = (12, 4))
    plt.subplots_adjust(wspace = 0.33)
    accuracy, time, method = [[], []], [[], []], [svm, gau]
    for i in f_select:
        for j, met in enumerate(method):
            new_X = skf.SequentialFeatureSelector(met, n_features_to_select = i).fit_transform(X, y)
            TRAIN_X, TEST_X, TRAIN_Y, TEST_Y = sms.train_test_split(new_X, y, test_size = 0.25, random_state = SEED)
            c, t = time_it(lambda: pcopy(clf).fit(TRAIN_X, TRAIN_Y))
            accuracy[j].append(c.score(TEST_X, TEST_Y))
            time[j].append(t)
    plt.subplot(1, 2, 1)
    plt.title("Accuracy obtained with less features.")
    plt.plot(f_select, accuracy[0], label = "Selection by svm")
    plt.plot(f_select, accuracy[1], label = "Selection by nb")
    plt.xlabel("feature_quantity")
    plt.ylabel("accuracy")
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.title("Time on fitting with less features.")
    plt.plot(f_select, time[0], label = "Selection by svm")
    plt.plot(f_select, time[1], label = "Selection by nb")
    plt.xlabel("feature_quantity")
    plt.ylabel("time")
    plt.legend()

fs_accuracy(best_model, ANIMALS_X, ANIMALS_Y)
```

De esto, podemos ver que a pesar de los esfuerzos realizados, aún no logramos que las estadísticas de nuestros algoritmos suban como lo esperado, es más, la realización de esta técnica presenta incluso disminuciones en lo que esperábamos tuviera un impacto positivo, por lo que podemos concluir que o la base de datos es muy compleja, o necesitamos otros métodos como por ejemplo, en vez de seleccionar características más bien reducir la dimensionalidad, tipo, aplicar técnicas de PCA y explorar los consiguientes resultados.
