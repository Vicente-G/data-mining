**Aprendizaje supervisado:** El aprendizaje supervisado es una colección de herramientas de aprendizaje de máquinas que se basa en la predicción a través de ejemplos.

Tenemos dos grandes tipos de algoritmos: la clasificación y la regresión. Su principal objetivo es que estos procesos se realicen de manera automática o semiautomática en base a datos que se tengan de manera histórica.

Estos procesos requieren inputs de información para poder hacer un proceso de comparación con los datos a analizar, en función de información histórica que se pueda obtener. Existe una profunda correspondencia entre los datos de entrada y los datos de salida. El aprendizaje supervisado depende de que el usuario actúe como una guía al algoritmo utilizado.

**Problema de clasificación:** Los problemas de clasificación son aquellos que a través de una tupla (x,y) donde x son los atributos de un objeto, pueden, a través de un clasificador, asignar un valor para el atributo y que es el cuál se intenta predecir en base al naturaleza de cada objeto. La clasificación se entrena con un conjunto tal que, en base a sus atributos se elije algún algoritmo de clasificación para que luego en un conjunto de prueba se pueda ver si es que el clasificador funcionó como era esperado o no.

La clasificación puede ser binaria, multiclase o multilabel. Siendo la clasificación binaria la más común de estas, ya que el valor de salida y sólo puede tomar 1 y sólo 1 de dos valores conocidos. En el caso de la clasificación multivariable y ∈ A tal que A es un conjunto cualquiera, de cardinalidad cualquiera. En el caso de la clasificación multilabel, la salida y puede ser distinta para elementos considerablemente similares. Para trabajar con este tipo de clasificación los algoritmos deben o transformarse (divide y vencerás; separar la base de datos en particiones) o adaptarse en función de la complejidad intrínseca de la base de datos a analizar. Algunos de los algoritmos más utilizados son: el vecino más cercano, que a través de métricas de distancia puede predecir la clase y de un objeto conociendo a los vecinos cercanos, support vector machine, algoritmo a través del cual utilizando un hiperplano (o recta o plano dependiendo de la dimensionalidad del problema) se puede hacer una separación de los objetos (Se entrará a detalle más adelante en la tarea).

## **1.2** Describa brevemente los siguientes clasificadores
- Árbol de decisión.
- Suport vector Machine (SVM) 
- Naive Bayes
- Vecinos mas ceranos (KNN)

**Árbol de decisión:** Los árboles de decisión son un algoritmo de clasificación tal que, para cada objeto recorre sus atributos y en función de las "respuestas" que el objeto dé, este puede clasificarlo en los distintos valores que toma y. Cada atributo del objeto será una pregunta y los posibles valores del atributo la respuesta, en correlación a esto, el árbol podrá ir descendiendo por los nodos hasta que en algún punto se encontrará con una hoja y por tanto, el clasificador entregará la clase del objeto.

**Suport vector Machine (SVM):** Como se mencionó brevemente en un inciso anterior, support vector machine crea un hiperplano para poder separar los distintos objetos y dependiendo de la posición en que queden respecto a este, se les podrá asignar una clase a cada uno. La idea es que este sea óptimo y por tanto, siempre maximizará las distancias respecto al objeto más cercano que tenga de cada clase. La distancia entre el objeto más cercano y el hiperplano será el margen de soporte, de ahí el nombre del modelo.

**Naive Bayes:** Conocida como la ley ingenua de Bayes, este algoritmo se basa en la fórmula de Bayes en función de que algo "pueda ser" probable a una cierta característica, en tanto, hace valoraciones $P(y_1|x), P(y_2|x)...P(y_n|x)$ y la probabilidad más alta es la que entrega el clasificador como clase para el objeto.

**Vecinos mas ceranos (KNN):** Se basa en métricas de distancias. Ubicado cierto objeto en el espacio se procede a ver las clases de los objetos cercanos a este (cantidad dada, elegida por el usuario), y en función de la naturaleza de sus "vecinos" se asigna una clase al objeto en cuestión.

## **1.3** Defina las siguientes métricas de desempeño
- Accuracy
- Falso positivo, falso negativo, verdadero positivo, verdadero negativo.
- Matriz de confusión
- Precisión y Recall 

**Accuracy:** Accuracy se refiere a exactitud y es una relación aritmética entre las predicciones correctas respecto a las predicciones hechas, se puede entender a través de la siguiente fórmula: 
$$Accuracy=\frac{VP+VN}{VP+VN+FP+FN}$$
tal que VP es verdadero positivo, VN verdadero negativo, FP falso positivo y FN falso negativo. Esta fórmula es una aplicación de la clásica fórmula de probabilidad. Accuracy puede ser muy poco fiable para grandes números, especialmente si una de las clases es pequeña.

**Falso positivo, falso negativo, verdadero positivo y verdadero negativo:** Entender esta métrica es considerablemente intuitivo, supongamos que un objeto debe ser clasificado de cierta forma y el algoritmo de clasificación es capaz de fallar. Luego, tenemos dos posibilidades, que el objeto por naturaleza sea positivo o negativo. Además, nuestro algoritmo debe clasificar en estas dos posibilidades. Luego, por cardinalidad tenemos $2^2$ posibles combinaciones de estas posibilidades tal vez, si el objeto es positivo y el catalogado positivo, obtenemos un verdadero positivo, en caso de ser positivo y que el clasificador diga que es negativo, tenemos un falso negativo, ya que no corresponde en tal clase. Análogamente, si un objeto es negativo y la máquina lo cataloga positivo, tenemos un falso positivo, y en caso de ser negativo y que el algoritmo le asigne correctamente su clase, tenemos un verdadero negativo.

**Matriz de confusión:** La matriz de confusión es una visión gráfica de los VP, VN, FP Y FN, en forma de dataframe, esta misma se utiliza para ver la cantidad de predicciones hechas en cada celda de la matriz, siendo el caso óptimo en el que los FP y FN son lo más cercanos a 0 posible. La matriz tiene características discretas y es muy útil para definir si un clasificador o el algoritmo de selección está funcionando de manera idónea.

**Presición y recall:** Ambas métricas se basan en los VP, VN, FP Y FN pero suelen ser más acertadas que accuracy, ambas son relaciones ariméticas de probabilidad respecto al funcionamiento del clasificador.
$$Presicion=\frac{VP}{VP+FP}$$

$$Recall=\frac{VP}{VP+FN}$$

Se puede apreciar que presición es más "ácida" que accuracy, ya que, por ejemplo, en casos donde los positivos son 98-99% este mismo podría tirar un accuracy muy alto, cosa que precisión podría revisar de manera más clara respecto al grado de fallos que tiene respecto el algoritmo. Con recall se aplica la misma idea. Es más, existe una métrica que combina ambas, cuya fórmula es:
$$F1=2*\frac{Recall*Presicion}{Recall+Presicion}$$

Cuya característica principal es que es muy sensible a outliers.

```python
random_tree = skt.DecisionTreeClassifier(splitter = "random", random_state = SEED)
best_tree = skt.DecisionTreeClassifier(splitter = "best", random_state = SEED)
linear_svm = sks.SVC(kernel = "linear", random_state = SEED)
poly_svm = sks.SVC(kernel = "poly", degree = 2, random_state = SEED)
gaussian_nb = skb.GaussianNB()
euclidean_knn = skn.KNeighborsClassifier(metric = "euclidean")
manhattan_knn = skn.KNeighborsClassifier(metric = "manhattan")
```

En este código seleccionamos distintos parámetros para los clasificadores estudiados, esto con el fin de posteriormente entrenarlos y compararlos. Esto considera las partes 3.2, 3.3, 3.4 y 3.5 del enunciado.

```python
def graph_model(model, attr, title):
    e_train, e_test = [], []
    for i in range(1, 61):
        model.set_params(**{attr : i})
        e_train.append(1 - model.fit(TRAIN, TRAIN_R).score(TRAIN, TRAIN_R))
        e_test.append(1 - model.fit(TRAIN, TRAIN_R).score(TEST, TEST_R))
    plt.title(title)
    plt.plot(list(range(1, len(e_train) + 1)), e_train, label = "Error de entrenamiento")
    plt.plot(list(range(1, len(e_test) + 1)), e_test, label = "Error de prueba")
    plt.xlabel(attr)
    plt.ylabel("Error")
    plt.legend()
    plt.show()

title = "Muestra de overfitting para clasificador Árbol de decisión."
graph_model(random_tree, "max_depth", title)
```

Tal como indica el gráfico, aquí podemos ver un caso típico de overfitting. En particular en las comparaciones realizadas más adelante, solo consideraremos los errores de prueba, puesto que como podemos observar, hay algunos clasificadores, en particular los árboles de decisión, que pueden fácilmente alcanzar una precisión perfecta sobre el conjunto de entrenamiento. Cabe mencionar de que como estamos en la búsqueda de un clasificador predictivo, toma muchísima más importancia la capacidad de obtener resultados certeros ante combinaciones desconocidas en $x$.

```python
def graph_comparison(model1, model2, attr, labels, scale = 1):
    model, y_model = [model1, model2], [[], []]
    best, best_prediction = 1, None
    for i in range(1, 61):
        for j in range(2):
            model[j].set_params(**{attr : i * scale})
            p_model = model[j].fit(TRAIN, TRAIN_R).predict(TEST)
            y_model[j].append(sum(p_model != TEST_R) / len(TEST_R))
            if ((sum(p_model != TEST_R) / len(TEST_R)) < best):
                best = (sum(p_model != TEST_R) / len(TEST_R))
                best_prediction = p_model
    plt.title(f"Errores al variar {attr} en un clasificador {labels[0]}.")
    x = np.linspace(1*scale, (len(y_model[0]) + 1)*scale, 60).tolist()
    plt.plot(x, y_model[0], label = labels[1])
    plt.plot(x, y_model[1], label = labels[2])
    plt.xlabel(attr)
    plt.ylabel("Error de prueba")
    plt.legend()
    return best_prediction

plt.figure(figsize = (18, 10))
plt.subplots_adjust(wspace = 0.2, hspace = 0.3)

plt.subplot(2, 2, 1)
labels = ["Árbol de decisión", "Mejor split", "Split aleatorio"]
tree_p = graph_comparison(best_tree, random_tree, "max_depth", labels)

plt.subplot(2, 2, 2)
labels = ["SVM", "Kernel lineal", "Kernel polinomial"]
svm_p = graph_comparison(linear_svm, poly_svm, "C", labels, scale = 0.05)

plt.subplot(2, 2, 3)
labels = ["Naive Bayes", "", "Gaussiano"]
bayes_p = graph_comparison(gaussian_nb, gaussian_nb, "var_smoothing", labels, scale = 0.02)

plt.subplot(2, 2, 4)
labels = ["KNN", "Distancia euclideana", "Distancia manhattan"]
knn_p = graph_comparison(euclidean_knn, manhattan_knn, "n_neighbors", labels)
```

# **Parte 4**

**Parte 4.1** Realice una comparación de los diferentes clasificadores en el conjunto de entrenamiento y de prueba. Para esto debe utilizar las métricas de desempeño estudiadas (en la sección 1.3).

```python
def metric_comparison(predictions, names):
    data = {
        "Precision" : [], "Recall" : [], "F1" : [], "Accuracy" : []
    }
    for p in predictions:
        report = skm.classification_report(p, TEST_R, output_dict = True)
        data["Precision"].append(report["1"]["precision"])     
        data["Recall"].append(report["1"]["recall"])
        data["F1"].append(report["1"]["f1-score"])
        data["Accuracy"].append(report["accuracy"])
    return pd.DataFrame(data, names)
    
predictions = [tree_p, svm_p, bayes_p, knn_p]
names = ["Árbol de decisión", "SVM", "Naive bayes", "KNN"]
metric_comparison(predictions, names)
```

**Parte 4.2** Realice un gráfico comparativo entre los clasificadores analizados y discuta las diferentes matrices de confusión.

```python
def metric_model(prediction, model_name = ""):
    conf_matrix = skm.confusion_matrix(prediction, TEST_R)
    names = {
        "xticklabels" : np.unique(prediction),
        "yticklabels" : np.unique(prediction)
    }
    sns.heatmap(conf_matrix, cbar = False, annot = True, **names)
    plt.title(f"Matriz de confusión para el modelo{' ' if model_name else ''}{model_name}.")
    plt.xlabel("Valor de verdad")
    plt.ylabel("Predicción")
    return conf_matrix

plt.figure(figsize = (20, 4))
plt.subplots_adjust(wspace = 0.33)

plt.subplot(1, 4, 1)
tree_m = metric_model(tree_p, "Árbol de decisión")

plt.subplot(1, 4, 2)
svm_m = metric_model(svm_p, "SVM")

plt.subplot(1, 4, 3)
bayes_m = metric_model(bayes_p, "Naive Bayes")

plt.subplot(1, 4, 4)
knn_m = metric_model(knn_p, "KNN")
```

```python
counters = [0, 0, 0, 0]
for i in list(range(7821, 31278382, 32126)):
    ex, px, ey, py = sms.train_test_split(SONAR_X,
                                          SONAR_Y,
                                          test_size = 0.3,
                                          random_state = i)
    clf = [
           skt.DecisionTreeClassifier(splitter = "best", max_depth = 10, random_state = i),
           sks.SVC(kernel = "poly", C = 3, degree = 2, random_state = i),
           skb.GaussianNB(),
           skn.KNeighborsClassifier(metric = "manhattan", n_neighbors = 3)
    ]
    scores = [j.fit(ex, ey).score(px, py) for j in clf]
    counters[scores.index(max(scores))] += 1

print(f"Árbol gana un {(counters[0] / sum(counters)) * 100}% de las veces.")
print(f"SVM gana un {(counters[1] / sum(counters)) * 100}% de las veces.")
print(f"Bayes gana un {(counters[2] / sum(counters)) * 100}% de las veces.")
print(f"KNN gana un {(counters[3] / sum(counters)) * 100}% de las veces.")
```
